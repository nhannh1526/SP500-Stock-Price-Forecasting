```python
import pickle
import pandas as pd

with open("results.pickle", "rb") as file:
    results = pickle.load(file)

results = pd.DataFrame.from_dict(results, orient="index")
index = results.index
models = ["Linear Regression", "Random Forest", "Ada Boost",
          "Light Gradient Boosting Machine", "Extreme Gradient Boosting", "ARIMA",
          "RNN", "LSTM", "GRU"]
```

```python
result_df = pd.DataFrame()

for model in models:
    df = pd.DataFrame(list(results[model]), index=index)
    df["Model"] = model
    result_df = pd.concat([result_df, df])

result_df.sort_index(inplace=True)
result_df.reset_index(inplace=True)
result_df.rename(columns={"index": "Symbol"}, inplace=True)
results = result_df.reindex(["Symbol", "Model", "RMSE", "MAE", "MAPE"], axis=1)
```

```python
results.groupby("Model").mean().sort_values("MAPE")
```

```python
results.groupby("Symbol").median().sort_values("MAPE")
```
